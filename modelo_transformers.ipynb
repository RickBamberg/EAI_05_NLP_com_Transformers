## Uso de Modelo Transformer Pré-Treinado
Neste notebook, vamos carregar um modelo Transformer pré-treinado (distilbert-base-uncased) da biblioteca Hugging Face. Vamos tokenizar uma frase de exemplo, alimentar o modelo com esses tokens e inspecionar a saída — os embeddings da última camada.

### Importações

```python
from transformers import AutoTokenizer, AutoModel
import torch
```

### Selecionando o modelo e tokenizer

```python
modelo_nome = 'distilbert-base-uncased'

tokenizer = AutoTokenizer.from_pretrained(modelo_nome)
modelo = AutoModel.from_pretrained(modelo_nome)

modelo.eval()  # Desativa o modo de treino
```

### Tokenizando uma frase de exemplo

```python
frase = "Transformers are powerful models for NLP tasks."
tokens = tokenizer(frase, return_tensors='pt')

print(tokens)
```

### Passando os tokens pelo modelo

```python
with torch.no_grad():
    saida = modelo(**tokens)
```

### Explorando a saída

```python
# Mostrar os tokens reais correspondentes à frase
tokens = tokenizer.tokenize(frase)
print("Tokens:", tokens)

for chave in saida:
    print(f"{chave}: {saida[chave].shape}")
```

### Visualizando os embeddings

```python
embeddings = saida.last_hidden_state
print("Shape dos embeddings:", embeddings.shape)

# Exemplo: vetor do primeiro token ([CLS])
vetor_cls = embeddings[0][0]
print("Vetor do token [CLS]:", vetor_cls[:10])  # Mostrando os 10 primeiros valores
print("Shape da saída (última camada):", saida.last_hidden_state.shape)
```

## Conclusão

Utilizando `AutoModel`, temos acesso direto à representação vetorial (embeddings) dos tokens. Isso nos permite criar modelos personalizados, extrair características e treinar redes para tarefas específicas.

Essa técnica é fundamental para tarefas de NLP como classificação, análise de sentimentos, extração de entidades, entre outras.
