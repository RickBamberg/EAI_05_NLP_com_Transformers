## 📘 Tokenização com Transformers (Hugging Face)
Neste notebook, você aprenderá os fundamentos de tokenização usando AutoTokenizer da biblioteca transformers. A tokenização é o primeiro passo essencial no pré-processamento de texto para modelos baseados em Transformers.

### 🔧 Importações

```python
from transformers import AutoTokenizer
```

### ✂️ Tokenização de frases

```python
tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased")

frase = "Transformers são poderosos para NLP!"
tokens = tokenizer.tokenize(frase)
ids = tokenizer.convert_tokens_to_ids(tokens)

print("Tokens:", tokens)
print("IDs:", ids)
```

### 📌 Explicação:

- A frase é dividida em sub-palavras usando o tokenizador do modelo DistilBERT.

- O método convert_tokens_to_ids transforma esses tokens em IDs inteiros que o modelo entende.

### 📚 Tokenização de múltiplas frases com Padding e Truncation

```python
frases = ["Estou aprendendo Transformers!", "Tokenização é a primeira etapa."]
encodings = tokenizer(frases, padding=True, truncation=True, return_tensors="pt")

print(encodings)
```

### 📌 Explicação:

- padding=True completa a sequência com zeros para igualar os tamanhos.

- truncation=True limita frases que ultrapassam o tamanho máximo.

- return_tensors="pt" retorna tensores do PyTorch.

### 🔄 Decodificação: IDs para texto

```python
ids = tokenizer.encode("Isso é incrível!", add_special_tokens=True)
print("IDs:", ids)
print("Texto:", tokenizer.decode(ids))
```

### 📌 Explicação:

- encode transforma o texto em IDs, incluindo tokens especiais ([CLS], [SEP]).

- decode converte os IDs de volta para texto legível.

### ✅ Conclusão
Neste notebook, você aprendeu:

- Como carregar um tokenizador pré-treinado com AutoTokenizer.

- Como tokenizar frases únicas e múltiplas.

- O uso de padding, truncation e tensores.

- Como converter IDs para texto.
