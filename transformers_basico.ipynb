## 1. Limitações dos Modelos Clássicos (RNNs e LSTMs)

Antes dos Transformers, tarefas de NLP eram resolvidas com redes recorrentes, como RNNs e LSTMs. Esses modelos processam sequências de forma sequencial, o que gera duas grandes limitações:

- **Dependência de longo prazo**: é difícil capturar relações distantes entre palavras em uma sequência.
- **Dificuldade de paralelização**: como processam token por token, é mais difícil treinar eficientemente em paralelo.

Essas limitações motivaram o desenvolvimento de arquiteturas que usam atenção em vez de recorrência.


## 2. A Arquitetura Transformer

O paper *Attention Is All You Need* (Vaswani et al., 2017) propôs uma nova arquitetura baseada inteiramente em atenção.

### Componentes principais:
- **Encoder**: transforma a entrada em representações contextuais.
- **Decoder**: gera a saída a partir das representações do encoder.
- **Atenção**: mecanismo que permite ao modelo focar em partes relevantes da entrada.

O grande avanço foi o uso da **Self-Attention**, onde cada palavra pode se relacionar com todas as outras da sequência ao mesmo tempo.


## 3. Entendendo Self-Attention (Simplificado)

Vamos fazer uma simulação do mecanismo de self-attention com vetores simples.


```python
import numpy as np

# Exemplo: 3 palavras com vetores de dimensão 4
X = np.array([
    [1, 0, 1, 0],  # palavra 1
    [0, 2, 0, 2],  # palavra 2
    [1, 1, 1, 1]   # palavra 3
])

# Matrizes de projeção (weights fictícios)
W_Q = np.random.rand(4, 4)
W_K = np.random.rand(4, 4)
W_V = np.random.rand(4, 4)

# Calculando Q, K, V
Q = X @ W_Q
K = X @ W_K
V = X @ W_V

# Calculando scores de atenção
scores = Q @ K.T
scores = scores / np.sqrt(K.shape[-1])  # escalonamento
weights = np.exp(scores) / np.exp(scores).sum(axis=-1, keepdims=True)  # softmax

# Resultado da atenção
output = weights @ V

print("Atenção aplicada:", output.round(2))
```

## 4. Positional Encoding

Transformers não têm ordem embutida como as RNNs. Para resolver isso, usam **Positional Encoding**, que adiciona informações sobre a posição dos tokens na sequência.

Esses encodings são somados aos embeddings das palavras.

### Fórmula usada:
Para cada posição \( pos \) e dimensão \( i \):

\[
PE(pos, 2i) = \sin\left(\frac{pos}{10000^{2i/d}}\right)
\quad\quad
PE(pos, 2i+1) = \cos\left(\frac{pos}{10000^{2i/d}}\right)
\]

Essa codificação permite que o modelo aprenda relações relativas e absolutas de posição.


## 5. Modelos baseados em Transformers

### 🔹 BERT
- Bidirecional
- Usado para classificação, QA, etc.

### 🔹 GPT
- Autoregressivo
- Usado para geração de texto

### 🔹 T5 / BART
- Encoder-Decoder
- Usado para tradução, resumo, etc.

Esses modelos são treinados em tarefas como preenchimento de máscara (BERT), geração de próxima palavra (GPT), ou tradução de sequência para sequência (T5).


## 6. Hugging Face na prática: Sentiment Analysis com `pipeline`

A Hugging Face oferece modelos prontos com interface simples para tarefas como classificação de texto, tradução, resumo, etc.


```python
from transformers import pipeline
import torch
```

```python
# Criando o pipeline de análise de sentimentos
classifier = pipeline(
    "sentiment-analysis",
    model="distilbert/distilbert-base-uncased-finetuned-sst-2-english"
)

# Testando com uma frase em português
result = classifier("Transformers revolucionaram o NLP!")
print("Transformers revolucionaram o NLP!")
print(result)
```
